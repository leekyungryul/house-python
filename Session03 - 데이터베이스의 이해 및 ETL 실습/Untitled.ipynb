```python
import requests, bs4
import pandas as pd
#웹페이지 html 소스 가져오기
resp = requests.get("https://sparkkorea.com/테스트/")
resp.encoding='utf-8'
html = resp.text
htmlobj = bs4.BeautifulSoup(html, 'html.parser')
tableTag = htmlobj.find(name='table',attrs = {"id":"test_table"})
```

```python
tableTag
```

```python
tbodyTag = tableTag.find(name='tbody')
```

```python
trs = tbodyTag.find_all(name='tr')
```
