21. 사원명을 출력하되 급여가 100만원 미만이면 0, 100만원~200만원 사이면 200, 200만원~300만원이면 300...(구간대를 1000만원까지) 을 함께 출력하시오(사원별 급여구간)

```python
def range(inValue):
    if inValue < 1000000:
        return 0
    if (inValue > 1000000) & (inValue <= 2000000):
        return 1000000
    if (inValue > 2000000) & (inValue <= 3000000):
        return 2000000
    if (inValue > 3000000) & (inValue <= 4000000):
        return 3000000
    if (inValue > 4000000) & (inValue <= 5000000):
        return 4000000
    if (inValue > 5000000) & (inValue <= 6000000):
        return 5000000
    if (inValue > 6000000) & (inValue <= 7000000):
        return 6000000
    if (inValue > 7000000) & (inValue <= 8000000):
        return 7000000
    if (inValue > 8000000) & (inValue <= 9000000):
        return 8000000
    if (inValue > 9000000) & (inValue <= 10000000):
        return 9000000
    else: return 10000000
```

```python
# selloutData['YEAR'] = selloutData.YEARWEEK.apply(year)
사원['급여구간'] = 사원.급여.apply(range)
```
