```python
import pandas as pd
import numpy as np
```

```python
selloutData = pd.read_csv('../dataset/kopo_product_volume.csv')
```

```python
selloutData.columns = selloutData.columns.str.lower()
```

```python
selloutData.head()
```

```python
def sub_function(data):
#     첫번째 ma값은 현재 volume값에 뒤의 행 2칸을 더한값의 평균값이된다.
    data['ma'] = data['volume'].rolling(window=5, center=True, min_periods=1).mean()
    return data
```

```python
sortedData = selloutData.sort_values(['regionid','productgroup','yearweek'])
sortedData.reset_index(inplace=True,drop=True)
sortedData.head()
```

```python
# 그룹을 묶어서 함수적용을 시키면 그룹간에 간섭을 받지 않게된다.
finalGroupResult2 = sortedData.groupby(['regionid','productgroup']).apply(sub_function)
finalGroupResult2.head()
finalGroupResult2.to_csv('./finalResult', index=False, encoding='ms949')
# 처리결과를 csv파일로 저장하기
# sortedData.to_csv('./middleResult', index=False, encoding= 'ms949')
```
