[데이터 통합] selloutData 자료에서 YEAR, WEEK 컬럼을 생성하고 WEEK 가 52 이하인 데이터만 조회한 후 refinedSelloutData 변수에 담으세요

```python
selloutData = pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
# selloutData
```

```python
# selloutData의 값들을 형변환 처리
selloutData = selloutData.astype({'REGIONID':str,
                                  'PRODUCT' : str,
                                  'YEARWEEK' : str,
                                  'QTY' : float})
```

```python
# yearweek의 year와 week를 분리하기 위해서 함수를 만들어봄
def year(inValue):
    outValue = inValue[0:4]
    return outValue

def week(inValue):
    outValue = inValue[4:6]
    return outValue
```

```python
# yearweek를 함수처리해서 새로운 컬럼에 나누어서 넣는다.
# selloutData['YEAR'] = selloutData.YEARWEEK.apply(year)
# selloutData['WEEK'] = selloutData.YEARWEEK.apply(week)
selloutData['YEAR'] = selloutData.YEARWEEK.str[0:4]
selloutData['WEEK'] = selloutData.YEARWEEK.str[4:6]
```

```python
selloutData
```

```python
refinedSelloutData = selloutData.loc[selloutData.WEEK.astype(int) <= 52]
refinedSelloutData
```

[불량 데이터 처리] kopo_channel_seasonality_new.csv 자료를 담은 selloutData 변수에서 QTY컬럼 음수(반품)인 경우 0, 양수인 경우 기존 QTY 값 유지하는 로직을 적용하여 QTY_NEW 컬럼을 추가하세요

```python
refinedSelloutData['QTY_NEW'] = \
    np.where(refinedSelloutData.QTY.astype(int) < 0, 0, refinedSelloutData.QTY)
```

```python
refinedSelloutData
```

refinedSelloutData 에서 → 지역, 상품, 연주차 컬럼순으로 오름차순 정렬하여 sortedData 변수에 담으세요

```python
# 먼저 정렬의 기준을 삼을 컬럼들을 리스트로 만들어서 변수에 담는다.
sortKey = ['REGIONID', 'PRODUCT', 'YEARWEEK']
# 위의 컬럼들을 기준으로 오름차순으로 변수에 담을수 있게 inplace는 false로 만들어준다.
sortedData = refinedSelloutData.sort_values(by = sortKey,
                                            inplace = False)
sortedData
# 만약 카피를 하지않고 sortedData = refinedSelloutData을 한다면 
# 둘다 같은 주소를 바라보기 때문에 서로연결되어서 각각 분리된 사용이 불가하다.
# sortedData = refinedSelloutData.copy()
```

[지역, 상품, 연도 별 집계] sortedData 에서 지역, 상품, 연도 단위 판매량(QTY_NEW) 의 평균 연산 후 groupData 변수에 담으세요 이후 컬럼명을 QTY_MEAN로 변경하세요

```python
# 지역,상품,연도별로 그룹정렬을 해야하므로 원하는 기준을 묶어서 리스트로 만들어준다.
groupkey = ['REGIONID','PRODUCT','YEAR']
groupData = sortedData.groupby(groupkey)[['QTY_NEW']].agg('mean')
groupData = groupData.rename(columns={'QTY_NEW' : 'QTY_MEAN'})
groupData.reset_index()
```

```python
groupkey = ['REGIONID','PRODUCT','YEAR']
sortedData.groupby(groupkey)["QTY"].describe()
```

이동집계함수

```python
# MA5라는 이름의 컬럼을 생성해서
# QTY를 기준으로 5개의 구간, 최소 1구간, 기준점은 중앙으로해서 이동집계를 구했다.
sortedData['MA5'] = sortedData.QTY.rolling\
                    (window = 5,
                    min_periods = 1,
                    center = True).mean()
```

```python
# sortedData
```

```python
# 최소구간을 설정하지않으면 5개구간을 충족하지 못하는 행은 nan으로 표시된다.
sortedData['MA5_BASIC'] = sortedData.QTY.rolling\
                    (window = 5,
#                     min_periods=1,
                    center=True).mean()
```

```python
# sortedData
```

```python
# center를 false로 두게되면 현재행을 기준으로 앞에서 부터 현재까지 계산한다. 
# 첫번째값은 본인값으로 나올것이다.
sortedData["MA5_BASIC_NO"] = sortedData.QTY.rolling\
                            (window = 5,
                            min_periods = 1,
                            center = False).mean()
```

```python
sortedData
```

```python
sortedData['EFFECT'] = sortedData.QTY_NEW / sortedData.MA5
```

```python
# sortedData
```

```python
# 처리결과를 csv파일로 저장하기
sortedData.to_csv('./middleResult', index=False, encoding= 'ms949')
```

```python
groupKey1 = ['REGIONID','PRODUCT','WEEK']
ratioData = sortedData.groupby(groupKey1)[['EFFECT']].agg('mean').reset_index()
# ratioData2 = sortedData.groupby(groupKey1)['EFFECT'].agg('mean')
```

#### inner조인

조인키 대상은 중복을 허락하지 않아야 한다.

중복이되는 조인키를 만들어보자

```python
# joinMaster에는 영향이 없도록 복사를해서 새로운 변수에 대입했다
testTable = joinMaster.copy()
testTable
```

```python
# regionid의 A01과 A02를 때어내어서 새로운 테이블로 만들고
# 새로운 변수명(testTable)에 대입하여 주었다.
# testTable = testTable.loc[0:1]
# testTable
```

```python
# testTable의 regionid컬럼의 A01을 A02로 바꾸어보자
# 중복되는 조인키를 만들어 보기 위해서 REGIONID가 A02인 데이터의 REGIONID를 A01로 변경
testTable.loc[testTable.REGIONID == "A02", 'REGIONID'] = "A01"
testTable
```

```python
# 그럼 중복된 조인키가 중복되어있는 testTable과 joinSellout을 inner조인 시켜보자
pd.merge(left=joinSellout,
        right=testTable,
        left_on="REGIONID",
        right_on="REGIONID",
        how="inner")
# inner join하였더니 joinSellout의 REGIONID가 A02인 데이터는 삭제되었다.
# 왜냐하면 testTable에는 조인키인 REGIONID에 A02가 없기 때문이다.
# 이러한 경우 right table에 동일한 조인키가 없더라도 left table의 데이터를 살리고 싶으면 
# left join을 하여야한다.
```

컬럼명이 다른경우 join

```python
# 컬럼명 변경하기
joinMaster = joinMaster.rename({'REGIONID':'AP2ID'},axis='columns')
joinMaster
```

```python
pd.merge(left=joinSellout,
        right=joinMaster,
        left_on="REGIONID",
        right_on="AP2ID",
        how='inner')
# 앞서 동일한 컬럼명을 사용했을때 join과 다른점은 
# 두 테이블의 조인키에 해당하는 컬럼명이 다르기 때문에 동일한 데이터를 가지는 
# 새로운 컬럼만 한줄 추가되었다. 그외에는 모두 동일하다.
```

#### 실습

새로운 테이블 생성

```python
columnList = ['REGIONID', 'PRODUCT', 'DATA', 'QTY']
data1 = pd.DataFrame([['서울', 'AA카드','20210512','3000'],
                      ['서울', 'AA카드','20210513','1000'],
                      ['서울', 'AA카드','20210514','2000'],
                      ['서울', 'AA카드','20210515','4000'],
                      ['서울', 'AA카드','20210516','5000']],columns = columnList)
```

```python
data1
```

기존 테이블을 복사해서 테이블 내용만 수정

```python
data2 = data1.copy()
data2.PRODUCT = 'BB카드'
data2
```

```python
data3 = data1.copy()
data3.REGIONID = '부산'
data3
```

```python
data4 = data3.copy()
data4.PRODUCT = 'BB카드'
data4
```

```python
mergedData = data1.append(data2).append(data3).append(data4)
# mergedData
```

프로모션 정보 테이블 생성

```python
promotionClumns = ['SUBSIDIARY','CARD_NAME','PROMOTION_NAME','PROMOTION_RATE']
promotinData = pd.DataFrame([
    ['서울','AA카드','서울_프로모션',0.5],
    ['부산','BB카드','부산_프로모션',0.3],
    ['부산','CC카드','부산_프로모션',0.25]],columns=promotionClumns)
promotinData
```

left table의 조인키가 right table에 중복되어 있는 경우

 - 아래와 같은 경우 부산이라는 데이터가 right table에 두종류가 있다.

```python
pd.merge(left=data4,
         right=promotinData,
         left_on='REGIONID',
         right_on='SUBSIDIARY',
         how='left')
# 조인키에 해당하는 컬럼명을 포함해서 컬럼명이 다른경우 
# 동일한 데이터를 가지고 있다하더라도 모두표시
```

위와 같은 경우를 예방하기 위해서 조인키에 두가지를 넣으면 된다.

조인키에 지역정보와 상품정보를 둘다 맵핑시키서 조인시키면 중복되지 않는다.

```python
pd.merge(left=data4,
        right=promotinData,
        left_on=['REGIONID', 'PRODUCT'],
        right_on=['SUBSIDIARY', 'CARD_NAME'],
        how='left')
# 부산 cc카드의 경우에는 조인되지 않는다.
```

병합시킨 카드정보전체를 프로모션 테이블과 조인시킨다.

```python
pd.merge(left=mergedData,
        right=promotinData,
        left_on=['REGIONID','PRODUCT'],
        right_on=['SUBSIDIARY','CARD_NAME'],
        how='left')
# 아래와 같은 경우 서울 bb카드와 부산 aa카드 같은 경우는 프로모션 테이블에 없으므로
# nan으로 표기된다.
```
