4. 생각해볼 문제 (지역, 상품, 분기별 평균 판매량을 산출하세요) 분기는 1 ~ 13,  14 ~ 26,  27 ~ 39,  40 ~ 52 로 구분 (53 제외)

```python
selloutData['WEEK'] = selloutData.YEARWEEK.astype(str).str[4:6]
```

```python
selloutData.head()
```

```python
# 지역,상품,주차별 평균판매량 
test = selloutData.groupby(['REGIONID','PRODUCT','WEEK'])['QTY'].mean().reset_index()
test.head()
```

```python
# 1분기 지역, 상품별 평균판매량
# 지역,상품,주차별 평균판매량에서 1분기만 빼서 지역,상품별 판매량 평균산출
test.loc[test.WEEK.astype(int) <= 13].groupby(['REGIONID','PRODUCT'])['QTY'].mean().reset_index()
```

```python
# 2분기 지역, 상품별 평균판매량
# 지역,상품,주차별 평균판매량에서 2분기만 빼서 지역,상품별 판매량 평균산출
test.loc[(test.WEEK.astype(int) > 13) & (test.WEEK.astype(int) <= 26)]\
    .groupby(['REGIONID','PRODUCT'])['QTY'].mean().reset_index()
```

```python
# 3분기 지역, 상품별 평균판매량
# 지역,상품,주차별 평균판매량에서 3분기만 빼서 지역,상품별 판매량 평균산출
test.loc[(test.WEEK.astype(int) > 26) & (test.WEEK.astype(int) <= 39)]\
    .groupby(['REGIONID','PRODUCT'])['QTY'].mean().reset_index()
```

```python
# 4분기 지역, 상품별 평균판매량
# 지역,상품,주차별 평균판매량에서 4분기만 빼서 지역,상품별 판매량 평균산출
test.loc[(test.WEEK.astype(int) > 39) & (test.WEEK.astype(int) <= 52)]\
    .groupby(['REGIONID','PRODUCT'])['QTY'].mean().reset_index()
```
