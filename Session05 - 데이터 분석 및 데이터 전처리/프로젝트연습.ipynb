```python
kopoData = pd.read_csv('../dataset/kopo_product_volume.csv')
```

```python
kopoDataA01 = kopoData.loc[kopoData.PRODUCTGROUP == 'ST0001']
```

```python
# 지역 상품 연주차별 정렬
sortKey = ['REGIONID','YEARWEEK']
koposortedData = kopoDataA01.sort_values(sortKey, ignore_index = True)
```

```python
koposortedData.head()
```

```python
koposortedData['MA5']=koposortedData['VOLUME'].rolling(window = 5, center = True, min_periods = 1).mean()
```

```python
koposortedData.head()
```

```python
koposortedData['EFFECT'] = round(koposortedData['VOLUME'] / koposortedData['MA5'] *100,2)
```

```python
koposortedData.head()
```

주차별 효과 산출

```python
koposortedData['WEEK'] = koposortedData.YEARWEEK.astype(str).str[4:6]
```

```python
groupKey = ['REGIONID','PRODUCTGROUP','WEEK']
weeklyEffect = koposortedData.groupby(by = groupKey)['EFFECT'].agg(['mean']).reset_index()
```
